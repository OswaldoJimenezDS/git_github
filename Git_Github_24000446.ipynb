# ¿Qué es git? 
### Control de versiones rápido

> Git es un sistema de control de versiones distribuido, gratuito y de código abierto, diseñado para manejar todo, desde proyectos pequeños hasta muy grandes con velocidad y eficiencia. - https://git-scm.com/

### Antecedentes y características
git es un un proyecto maduro y hoy cuenta con mantenimiento activo, desarrollado originalmente por Linus Tolvalds, famoso por la creación del kernel del sistema operativo Linux (1991). git se basa en una arquitectura distribuidad DVCS (Distributed version control system) esto quiere decir que en lugar de tener un [unico lugar para el historial de versiones, como ocurre con los predecesores CVS o SVN, la copia de trabajo del c[odigo de cada desarrollador es también un repositorio que puede contener el historial completo de los cambios realizados.

git no solo ofrece una arquitectura distribuida, también cuenta con características dalta seguridad, flexibilidad y buen rendimiento.

#### Rendimiento
Una de las características importantes de git es el rendimeinto derivado de nuevos cambios, la ramificación, la fusión y comparación de versiones han hecho que el redimiento sea bien calificada comparada con otras alternativas. Esta característica se logra por medio del manejo de los archivos, ya que git no se basa en el nombre de los archivos, ya que esto puede resultar contraproducente al este ser suceptible a cambios, en su lugar se bas en el contenido de los archivos y con ello logra estructurar la mejor dorma de estructurar el almacenamiento y el árbol de archivos; git emplea una cambinación de codificación delta(almacena las diferencias de código) y compresión guardando explícitamente el contenido de los directorios y metadatos de las versiones.
#### Seguridad
La principal prioridad de git es conservar la integridad del código fuente que gestiona, así como el contenido de los archivos, las verdaderas relaciones entre estos y los directorios, todo desde las versiones, etiquetas y confirmaciones forman parte del repositorio de git, los cuáles están protegidos por un algoritmo de hash criptográfico seguro (SHA1), de esta forma se guarda y protege el historial y el código frente a modificaciones accidentales o maliciosas y se garantiza trazabilidad.
#### Flexibilidad
Mantener un equilibrio entre segudidad, rendimeinto y flexibilidad es un tema complejo, pero git dentro de sus objetivos combina estos elemento de forma muy eficiente, para ello cuenta con la capacidad de manejar varios tipo de flujos de trabajo de desarrollo no lineal, en su eficiencia en proyectos grandes, medianos y pequeños ya la compatibilidad con varios sistemas y protocolos. git ofrece ademas la posibiliad de ramificación y etiquetado adicional al almacenamiento de historial lo que ofrece un gran nivel de seguimiento.


### Aspectos básicos
#### Ramas
Cada desarrollador puede gardar sus cambios en su repositorio local, esto da como resultado que existan diferentes cambios basados en una misma confirmación (Para que los cambios sean aceptados por git deben ser guardados con una confirmación), para ello git proporciona una herramienta para aislar los archivos y volver a cambiarlos posteriormente. Las ramas son punteros muy ligeros para el trabajo en curso los cuales administran esta separación, una vez terminado el trabajo se puede volver a la rama principal del equipo.
 
> https://learn.microsoft.com/es-es/devops/develop/git/what-is-git
#### confirmaciones y archivos
Los estados en los que git mantiene los archivos son: modificados, almacenados provisionalmente y confirmados. Cuando un archivo es modificado por primera vez estos cambios solo existen en el directorio de trabajo, aún no forman parte de una confirmación de modificación ni en el historial. Los archivos serán almacenados almacenados temporalmente por el desarrollador, el área de almacenamiento provisional contendrá los cambios que se incluirán en la siguiente confirmación, cuando se esté satisfecho con los cambios los archivos se empaquetarán como una confirmación con un mesaje descriptivo para los cambios la cuál pasa a formar parte del historial de desarrollo.

> https://git-scm.com/book/en/v2/Git-Basics-Recording-Changes-to-the-Repository

### Comandos básico en git

1. git init: Este comando se utiliza para iniciar un nuevo repositorio Git en un directorio. Marca el comienzo del rastreo de cambios en ese proyecto.
   
3. git clone: Permite la clonación de un repositorio existente. Es útil cuando se desea obtener una copia local de un proyecto remoto, ya sea propio o de un compañero.
 
4. git add: Este comando añade cambios al área de preparación (staging area) en preparación para realizar un commit, puede utilizarse para agregar archivos específicos o todos los archivos modificados.

5. git commit: Captura un punto en la historia del proyecto. Se utiliza para confirmar los cambios realizados y proporciona un mensaje descriptivo para documentar esos cambios.
   
6. git status: Proporciona información sobre el estado actual del repositorio. Indica qué archivos han sido modificados, cuáles están en el área de preparación y si hay cambios pendientes de confirmar.

7. git branch: Muestra todas las ramas presentes en el repositorio. Puede ser utilizado para crear nuevas ramas o cambiar entre ellas.

8. git merge: Fusiona los cambios de una rama a otra. Es esencial para integrar características o correcciones de errores de una rama de desarrollo a la rama principal.

9. git pull: Recupera los cambios desde un repositorio remoto y los fusiona automáticamente con el repositorio local.

10. git push: Envía los cambios locales al repositorio remoto. Es necesario para compartir el trabajo con otros miembros del equipo.

11. git log: Muestra un historial detallado de los commits realizados en el repositorio, incluyendo el autor, la fecha y el mensaje de confirmación.

# ¿Qué es github?

github es una plataforma de desarrollo colaborativo (forja), que permite alojar proyectos utilizando git para el control de versiones, control de acceso, seguimiento de errores, solicitudes de gestiones de software, gestión de tareas, integración continua y wikis para cada uno de los proyectos. 

github tiene su sede en California y ha sido subsidiaria de Microsoft desde 2018 cuando esta fué adquirida por USD 7,500 millones. En un inicio algunos vieron con preocupación la compra por lo que algunas cuentas se dieron de baja aunque no fueron significativas, en Enero de 2023 se reportaban 100 millones de desarrolladores y mas de 372 millones de repositorios de los cuales 28 millones eran públicos.

La plataforma de github estaba originalmente desarrollada en Ruby on Rails, con el paso del tiempo esto ha cambiado y actualemente está basada en un software de código abierto denominado "Git", dentro de ella es posible almacenar, compartir y trabajar junto a otros usuarios para escribir código. github implementa repositorios para almacenar los proyectos y sobre ellos se permite:
1. Presentar o compartir el trabajo.
2. Seguir y administrar los cambios a lo largo del tiempo.
3. Dejar que otros usuarios revisen el código.
4. Colaborar en un proyecto compartido




### ¿Cómo se cambinan git y github?
Al cargar archivos en github, estos son almacenados en un repositorio de git, lo que implica que al realizar cambios o confirmaciones de cambio en los archivos de github, estos tendrán un seguimiento y los cambios serán administrador por git, muchas actividades de git pueden ser complementadas por github desde un navegador.

Muchos desarrolladores trabajan con git en sus repositorios locales para luego sincronizan continuamente los cambios con github para ello se usa un repositorio remoto, en la actualidad existen muchas herramientas para hacerlo, una de ellas es githb desktop. github ayuda a administrar el flujo a través de las solicitudes de cambio.
